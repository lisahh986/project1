```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
# loading data
life_data_path = "fuzzy.csv"
life_df = pd.read_csv(life_data_path)
# print(life_df.head())


life_df.drop(columns=life_df.columns[0], axis=1, inplace=True)
```

```python
# data cleaning, this might be too aggressive. (removed U.S.A.)
# Maybe we extract only relevant columns, then do a dropna to hopefully preserve some data. 
clean_life_df = life_df.dropna() 
```
